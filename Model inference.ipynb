```python
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np
from google.colab import drive
drive.mount('/content/drive')


# 1. Load your fine-tuned model
trained_model = SentenceTransformer('/content/drive/MyDrive/volunteer_model_v1')

# 2. A complex need from the field
new_need = ["Recent floods have damaged the local school and contaminated the well. We need help with repairs and water safety."]

# 3. Your Kaggle Volunteer Pool
volunteers = [
    "Nurse with emergency triage experience",
    "Civil Engineer specializing in water systems",
    "Primary School Teacher",
    "Digital Marketing Specialist"
]

# 4. Generate Embeddings
need_emb = trained_model.encode(new_need)
vol_embs = trained_model.encode(volunteers)

# 5. Calculate Match
scores = cosine_similarity(need_emb, vol_embs)[0]

print("--- Smart Match Results ---")
for i in np.argsort(scores)[::-1]:
    print(f"{volunteers[i]}: {scores[i]:.4f}")
```
